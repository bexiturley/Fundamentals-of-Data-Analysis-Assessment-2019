# amount that restaurant staff receive in tips based on various indicator data using Seaborn's Facetgrid.  Linespace makes the numbers evenly spaced.

```python
tips['tip_pct'] = 100 * tips['tip'] / tips['total_bill']

grid = sns.FacetGrid(tips, row="sex", col="time", margin_titles=True)
grid.map(plt.hist, "tip_pct", bins=np.linspace(0, 35, 35), color="r" );
```

# The above graphs show the tip percentage broken down between male and female for both lunch and dinner time.  Lunch time tips are broadly similar with both the male and femals.  Dinner time is quite different with the male giving far greater tips. 

```python
tips['tip_pct'] = 100 * tips['tip'] / tips['total_bill']

grid = sns.FacetGrid(tips, row="day", col="smoker", margin_titles=True)
grid.map(plt.hist, "tip_pct", bins=np.linspace(0, 35, 35), color="b" );
```
