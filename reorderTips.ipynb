# Regression

Linear regression predicts relationships between variables using statistics.  Fitting lines to datasets.  One of the easier to understand is linear regression.  Two related variables are taken and we try to find a linear function to estimate and explain the relationship between one dependent variable and one or more independent variables.

```python
# Plot data and a linear regression model fit.
sns.relplot(x="total_bill", y="tip", data=tips);
```

```python
sns.regplot(x="total_bill", y="tip", data=tips);
```

```python
# Below is the same data but this time implot is used.  Implot uses regplot and facetgrid.  
sns.lmplot(x="total_bill", y="tip", data=tips);
```

The three plots show the exact same information but the last two have a grid and creates a line going through it making it easier to visualise the information and see if the data is a fit for a linear regression model.

There appears to be some correlation between tips and total bill as there are no outlying values in the upper triangle while the lower triangle does have a few but over all the greatest concentration seems to be going in a line and does not appear to be totally random.

```python
# Regression plots using FacetGrid broken into the individual days
g = sns.FacetGrid(df, col="day") 
g.map(sns.regplot, "total_bill", "tip") 
plt.xlim(0, 55) 
plt.ylim(0, 10)
```

On Thursday the values are more clustered towards the bottom of the grid meaning the total bill cost was lower and the tips followed suit.  Friday doesnt have as many data points but the ones that are there fit the line better.  Saturday shows a lot of activity in the bottom half with most points congretrating to the line or nearby but a few of the tips and total bill points are on the higher side and appear at the top right corner of the plot.  Sunday has the greatest spread away from the mean but the points stay more towards the middle with less on the bottom and the top of the line.  

Looking at the graps it appears that there were more points on Saturday.  The less amounts stay fairly close to the line but as the amount on the total bill increases so does the value of the tips.  It would appear that Saturday would be the best day to work in the restaurant for higher tips but that doesnt say whethere it was lunch time or dinner time.  It is also fair to say that the information is skewed as there was no lunch tips recorded Saturday or Sunday.  And there was no Thursday dinner data recorded for males.

```python
# Using seaborn's facetgrid looking at tip percentage for different serving times for each sex.  linespace evenly spaces the numbers.
tips['tip_pct'] = 100 * tips['tip'] / tips['total_bill']

grid = sns.FacetGrid(tips, row="sex", col="time", margin_titles=True)
grid.map(plt.hist, "tip_pct", bins=np.linspace(0, 35, 35), color="r" );
```

 Lunch time tips are broadly similar with both the male and femals.  Dinner time is quite different with the male giving far greater tips. 

```python
# Tip percentage for the different meal times for each day this time looking at whether the customers were smokers or not.
tips['tip_pct'] = 100 * tips['tip'] / tips['total_bill']

grid = sns.FacetGrid(tips, row="day", col="smoker", margin_titles=True)
grid.map(plt.hist, "tip_pct", bins=np.linspace(0, 35, 35), color="b" );
```

```python
# A KDE plot is a lot like a histogram, it estimates the probability density of a continuous variable.
plt.figure(figsize=(10,6))
sns.kdeplot(tips['total_bill'], label='Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Probability Density')
plt.title('Probability density plot of the total bill')
plt.show()
```

```python
# A KDE plot is a lot like a histogram, it estimates the probability density of a continuous variable.
plt.figure(figsize=(10,6))
sns.kdeplot(tips['tip'], label='Tip')
plt.xlabel('Tip')
plt.ylabel('Probability Density')
plt.title('Probability density plot of the tips')
plt.show()
```

```python
# kernel density estimation in 2 dimensions, we can do this with seaborn also. KDE for total bill vs tip plotted on 2 dimension plot. Also overlayed this plot with the scatter plot so the outliners are visible. 
sns.kdeplot(tips['total_bill'], tips['tip'], cmap='magma_r')
plt.scatter(tips['total_bill'], tips['tip'], marker='x', color='r', alpha=0.5)
plt.xlabel('total bill')
plt.ylabel('tip')
plt.title('KDE plot of total bill vs tips')
plt.show()
```
