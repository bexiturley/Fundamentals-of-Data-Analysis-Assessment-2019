# Regression

Linear regression predicts relationships between variables using statistics.  Fitting lines to datasets.  One of the easier to understand is linear regression.  Two related variables are taken and we try to find a linear function to estimate and explain the relationship between one dependent variable and one or more independent variables.

```python
# Plot data and a linear regression model fit.
sns.relplot(x="total_bill", y="tip", data=tips);
```

```python
sns.regplot(x="total_bill", y="tip", data=tips);
```

```python
# Below is the same data but this time implot is used.  Implot uses regplot and facetgrid.  
sns.lmplot(x="total_bill", y="tip", data=tips);
```

The three plots show the exact same information but the last two have a grid and creates a line going through it making it easier to visualise the information and see if the data is a fit for a linear regression model.

There appears to be some correlation between tips and total bill as there are no outlying values in the upper triangle while the lower triangle does have a few but over all the greatest concentration seems to be going in a line and does not appear to be totally random.

```python
# Using box plots to look at the total bill but able to see the minimum, first quartile, median, third quartile, and maximum. Shows outliers and what their values are
sns.set(style="whitegrid")

tips = sns.load_dataset("tips")
sns.boxplot(y=tips["total_bill"])
```

Here we can see that the median total bill spent was just below 20.  The least amount was just above zero and the max was just above 50.  The shaded blue area begins with the first quartile.  The black line is the median and the top of the blue box is the third quartile. The median separates the upper half from the lower half. If all the values withing the dataset are laidout in order the one in the middle is the median.  The first quartile is the median of the values before the median detailed in the lines above here. The third quartile is the median of the values of the area above the median.  The interquartile Range begins at the first quartile and ends at the third quartile, taking in the median.  The diamond shapes above the upper adjacent value are know as an outliner.

```python
# A KDE plot is a lot like a histogram, it estimates the probability density of a continuous variable.
plt.figure(figsize=(10,6))
sns.kdeplot(tips['total_bill'], label='Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Probability Density')
plt.title('Probability density plot of the total bill')
plt.show()
```

From the kernel density plot we can see the graph peaks at approximately 15 dollars.  This signifies that most of the bills tend to cost this much.

```python
# A KDE plot is a lot like a histogram, it estimates the probability density of a continuous variable.
plt.figure(figsize=(10,6))
sns.kdeplot(tips['tip'], label='Tip')
plt.xlabel('Tip')
plt.ylabel('Probability Density')
plt.title('Probability density plot of the tips')
plt.show()
```

In this KDE the peak occurs at 2 and begins to decrease rapidly once 4 dollars is reached. This would indicate that most of the tips are between this range.
