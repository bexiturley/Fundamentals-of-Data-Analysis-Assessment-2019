```python
# Begin to take a deeper look at the information within tips dataset. How many males vs females
tips.groupby(by='sex').size()
```

From the above infromation, 157 meals were served to male diners and 87 to female diners.  That is almost twice as many males as females.  

```python
# What was the average of the total bill for lunch and dinner 
tips.groupby(by='time')['total_bill'].mean()
```

The average cost of a lunch was 17.17 and the average cost of each dinner served was 20.80.

```python
# What was the average of the total bill broken down into males and females
tips.groupby(by='sex')['total_bill'].mean()
```

The average male diner spent 20.75 while the average female spent 18.06

```python
# Grouped according to the sex of the diners, looking at the total bill column.  The describe() function gives more statistics than the one used above
tips.groupby(by="sex")["total_bill"].describe()
```

The count refers to the number of valid values within the table. So there are no details missing.  The mean is the sum of all the values within the column devided by the count. As already written above, the mean for males are 20.75 and females are 18.06. The standard deviation is a measure of how far away from the mean values typically are. A big standard deviation indicated the spread away from the mean the values are. The min value is the minimum abount which was spent while the max means the opposite. The 50% is the median, half the values are above it and half below it. Here we see that males consistently spent more than females for all the percentages listed above. 

```python
# How many meals were served to males and females subdivided by days of the week.  The by argument, used with the groupby, and a list of the columns we want to group by.
tips.groupby(by=['sex', 'day']).size()
```

From the above Thursday and Friday meals remained broadly similary when broken down by sex. But for Saturday and Sunday there is a huge discrepancy with many more males getting served.

```python
# The max daily tips
tips.groupby(by='day')['tip'].max() 
```

The max daily tips were almost the same Thursday and Sunday with a dip on Friday and a large increase on Saturday.

```python
# The max daily tips further divided by sex
tips.groupby(by=['sex', 'day']).max() 
```

Males gave higher tips when compared to their female counterparts for 3 out of the 4 days. 

```python
# The smallest daily tip the server received on each day of the week by each sex.
tips.groupby(by=['sex', 'day']).min() 
```

Females gave the smallest tip every day when compared to the males.

```python
# Total daily spend by each sex
tips.groupby(by=['sex', 'day'])['total_bill'].sum()
```

Males spent more each day and significantly more than females at the weekends. Fridays totals are way lower than the rest. 

```python
# Daily tip total by each sex
tips.groupby(by=['sex', 'day'])['tip'].sum()
```

Again, the total of males tips were higher each day. A big dip for both on Friday and males significantly larger at the weekend than the females.

```python
# The total spent each meal time, each day by each sex
tips.groupby(by=['sex', "time", 'day'])['total_bill'].sum()
```

There is no data for lunch on Saturday or Sunday.  Males spent more on lunch on Thursday with females spending more on Friday but the figure is way lower than Thursday.  As there is only two days of lunch spends and tips, and with such a discrepancy in total amount in spent, I would think it is not good enough to model data on by itself.  There are no dinner figures for males on Thursday.  And only a small amount spent on Thursday by females.  Males spent more on the remaining days.   

```python
# Total daily tip, by each sex for each meal
tips.groupby(by=['sex', "time", 'day'])['tip'].sum()
```

Males gave more tips for lunch on Thursday but not Friday.   There are no figures for males on Thursday dinner.  But for the remining dinner days they outtipped the women.

```python
# Using seaborn's facetgrid looking at tip percentage for different serving times for each sex.  linespace evenly spaces the numbers.
tips['tip_pct'] = 100 * tips['tip'] / tips['total_bill']

grid = sns.FacetGrid(tips, row="sex", col="time", margin_titles=True)
grid.map(plt.hist, "tip_pct", bins=np.linspace(0, 35, 35), color="r" );
```

 Lunch time tips are broadly similar with both the male and femals.  Dinner time is quite different with the male giving far greater tips. 

```python
# Tip percentage for the different meal times for each day this time looking at whether the customers were smokers or not.
tips['tip_pct'] = 100 * tips['tip'] / tips['total_bill']

grid = sns.FacetGrid(tips, row="day", col="smoker", margin_titles=True)
grid.map(plt.hist, "tip_pct", bins=np.linspace(0, 35, 35), color="b" );
```

```python
# A KDE plot is a lot like a histogram, it estimates the probability density of a continuous variable.
plt.figure(figsize=(10,6))
sns.kdeplot(tips['total_bill'], label='Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Probability Density')
plt.title('Probability density plot of the total bill')
plt.show()
```

```python
# A KDE plot is a lot like a histogram, it estimates the probability density of a continuous variable.
plt.figure(figsize=(10,6))
sns.kdeplot(tips['tip'], label='Tip')
plt.xlabel('Tip')
plt.ylabel('Probability Density')
plt.title('Probability density plot of the tips')
plt.show()
```

```python
# kernel density estimation in 2 dimensions, we can do this with seaborn also. KDE for total bill vs tip plotted on 2 dimension plot. Also overlayed this plot with the scatter plot so the outliners are visible. 
sns.kdeplot(tips['total_bill'], tips['tip'], cmap='magma_r')
plt.scatter(tips['total_bill'], tips['tip'], marker='x', color='r', alpha=0.5)
plt.xlabel('total bill')
plt.ylabel('tip')
plt.title('KDE plot of total bill vs tips')
plt.show()
```
